```python
# 1. Explain the key features of Python that make it a popular choice for programming

'''
Python is one of the most popular programming languages today because of several key features that make it beginner-friendly, powerful, and versatile. Here are the main reasons why Python is widely used:
    1. Python uses readable, English-like syntax, which makes it easy for beginners to learn and understand.
    2. Python code is executed line-by-line, which makes debugging and testing easier. it does not need to compile a compile a code before run it.
    3. it is a dynamically typed language, it means in python no need to declare data type explicitly
    4. it supports both oops and functional programming
    5. Python community is very wide, so it's easy to find tutorial, documentation and support
    6. Python provides a wide range of built-in modules and libraries for everything from math to web development to data analysis like Numpy, pandas, tensorflow, flask, etc.

'''

# 2.  Describe the role of predefined keywords in Python and provide examples of how they are used in a program

'''
Predefined keywords in Python are reserved words that have special meaning and functionality in the language. 
We cannot use them as variable names, function names, or identifiers because they are part of the Python syntax. some of the mostly used keywords in python are: if, else, elif, for, in, range, def, return,  etc

examples for:

if, else, elif - conditional statement
'''
age = 18
if age == 18:
    print("eligible for vote")
elif age >18
    print("eligible for vote
else:
    print("Not eligible for vote")
# output : eligible for vote

'''
for, in, range – Loops
'''
for i in range(5):
    print(i)

''' output  : 0 1 2 3 4 '''

'''
def, return – Function Definition
'''
def add(a, b):
    return a + b

print(add(3, 4))  

''' Output: 7 '''

# 3  Compare and contrast mutable and immutable objects in Python with examples

'''
Mutable objects are the objects that can be changed after creation whereas Immutable objects are Objects that cannot be changed once created.
list, dict,set,bytearray are the examples of mutable objects whereas 	int, float, str, tuple, bool, bytes are the examples of immutable objects

Example of mutable List:
'''
my_list = [1, 2, 3]

my_list.append(4)
print(my_list)         # Output: [1, 2, 3, 4]

'''  in this example it can be clearly seen that after creation my_list object we are able to modify the object '''
 ''' Immutable example '''

name = ('t','a','s')
name[1] = 'b'

"""
---------------------------------------------------------------------------
TypeError                                 Traceback (most recent call last)
Cell In[11], line 2
      1 name = ('t','a','s')
----> 2 name[1] = 'b'

TypeError: 'tuple' object does not support item assignment
"""

# 4 Discuss the different types of operators in Python and provide examples of how they are used

'''
Arithmetic Operators: it is used to do basic mathematics

'''
a = 10
b = 5

print(a + b)  # Addition → 15
print(a - b)  # Subtraction → 5
print(a * b)  # Multiplication → 50
print(a / b)  # Division → 2.0
print(a % b)  # Remainder → 0
print(a ** 2) # Power → 100

'''  Assignment Operators : it is used to give or update values of variables '''
x = 5       # Assign 5 to x
x += 2      # Same as x = x + 2 → x becomes 7
x *= 3      # x = x * 3 → x becomes 21

'''  Comparison Operators: it is Used to compare two things   '''
a = 10
b = 5

print(a == b)  # Is a equal to b? → False
print(a > b)   # Is a greater than b? → True
print(a != b)  # Are a and b not equal? → True

''' Logical Operators :  Used with conditions '''
x = 10

print(x > 5 and x < 15)  # True (both are true)
print(x > 20 or x < 15)  # True (one is true)
print(not x > 5)         # False (x > 5 is true, but we reversed it)

'''  Membership Operators : Check if item is present '''

fruits = ['apple', 'banana']

print('apple' in fruits)     # True
print('mango' not in fruits) # True

''' Identity Operators: Check if same object '''
x = [1, 2]
y = x
z = [1, 2]

print(x is y)     # True (same object)
print(x is z)     # False (same content, but different objects)

''' Bitwise Operators : Work on bits '''

a = 5  # 101 in binary
b = 3  # 011 in binary

print(a & b)  # AND → 1 (001)
print(a | b)  # OR → 7 (111)

# 5  Explain the concept of type casting in Python with examples

''' 
Type casting means changing the data type of a value from one type to another.

For example:

 1. Convert a string to an integer

 2. Convert a float to an integer

 3. Convert an integer to a string

There are two type of type casting, implicit and explicit type casting
Implicit Type Casting : Python automatically changes the type when needed.
example : 
'''
a = 5       # int
b = 2.0     # float

c = a + b   # int + float = float
print(c)    # 7.0
print(type(c))  # <class 'float'>

''' Python changed the result to float automatically. '''

'''  Explicit Type Casting :  we manually change the type using functions like:

int() – to convert to integer

float() – to convert to float

str() – to convert to string

list() – to convert to list

here are practical examples:
'''
'''Convert String to Integer'''

x = "10"
y = int(x)   # Now y is 10 (int)
print(y + 5) # 15

''' Convert Integer to Float'''

a = 7
b = float(a)
print(b)     # 7.0

''' Convert Integer to String '''

num = 100
text = str(num)
print(text + " is a number")  # "100 is a number"

''' Convert List to Set '''
my_list = [1, 2, 3, 1]
my_set = set(my_list)
print(my_set)   # {1, 2, 3}

''' but  not all conversion works directly '''
s = "abc"
num = int(s)   #  since abc cannot be converted into number equivalent so This will give an error: ValueError


# 6 How do conditional statements work in Python? Illustrate with examples

'''
Conditional statements let us run different code based on different conditions.
They help the program make decisions.

Here are basic Types of Conditional Statements:

if

if...else

if...elif...else

Nested if (if inside another if)

now i am explaining all of the one by one with practical example

1. if statement : it runs the code only if the condition is True.
'''
age = 18

if age >= 18:
    print("You are eligible to vote")

''' 2. if...else statement : it runs one block if condition is True, another block if False. '''

age = 16

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")

''' 3. if...elif...else statement: it is used to check multiple conditions. '''

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")

'''  4. Nested if statement: An if inside another if. '''
x = 10

if x > 5:
    if x < 15:
        print("x is between 5 and 15")


#7  Describe the different types of loops in Python and their use cases with examples

''' 
A loop is used to repeat a block of code multiple times.
types of loop : for loop, while loop, nested loops


1.  for loop
Used to loop through a sequence (like a list, string, range, etc.)

'''
fruits = ['apple', 'banana', 'mango']
for fruit in fruits:
    print(fruit)

# it will print every elment present in fruits , 
# apple
# banana
# mango

''' 2. while loop : Repeats a block as long as the condition is true. '''

i = 1
while i <= 5:
    print(i)
    i += 1


# This will print numbers 1 to 5.


''' 3. Nested Loops : A loop inside another loop. '''

for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j}")

'''
output :
i=1, j=1
i=1, j=2
i=2, j=1
i=2, j=2
i=3, j=1
i=3, j=2
'''
```
